```python
import cv2
import matplotlib.pyplot as plt
```

```python
face_model = cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")
```

```python
cam = cv2.VideoCapture(0)
```

```python
_ , img_webcam = cam.read()
```

```python
def ishow(image):
    cv2.imshow('image', img_webcam)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
ishow(img_webcam)
```

```python
gray = cv2.cvtColor(img_webcam,cv2.COLOR_BGR2GRAY)
```

```python
face = face_model.detectMultiScale(
    gray
)
```

```python
face
```

```python
plt.imshow(gray)
```

```python
x = face[0][0]
y = face[0][1]
x2 = x + face[0][2]
y2 = y + face[0][3]
```

```python
rec = cv2.rectangle(img_webcam, (x,y), (x2,y2), (0,250,0), 3)
ishow(rec)
```

```python
plt.imshow(rec)
```
