```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("arefe.jpg")
```

```python
def ishow(image):
    cv2.imshow('image', img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
ishow(img)
```

```python
face_model = cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")
```

```python
left_eye = cv2.CascadeClassifier("haarcascade_lefteye_2splits.xml")
```

```python
right_eye = cv2.CascadeClassifier("haarcascade_righteye_2splits.xml")
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
ishow(gray)
```

```python
face = face_model.detectMultiScale(
    gray
)
```

```python
face
```

```python
eye1 = left_eye.detectMultiScale(
    gray
)
```

```python
eye1
```

```python
eye2 = right_eye.detectMultiScale(
    gray
)
```

```python
eye2
```

```python
plt.imshow(gray)
```

```python
x = face[0][0]
y = face[0][1]
x2 = x + face[0][2]
y2 = y + face[0][3]
```

```python
xa = eye1[0][0]
ya = eye1[0][1]
x1 = xa + eye1[0][2]
y1 = ya + eye1[0][3]
```

```python
xb = eye2[0][0]
yb = eye2[0][1]
x3 = xb + eye2[0][2]
y3 = yb + eye2[0][3]
```

```python
rec = cv2.rectangle(img, (x,y), (x2,y2), (0,250,0), 3)
```

```python
rec1 = cv2.rectangle(img, (xa,ya), (x1,y1), (250,0,250),2)
```

```python
rec2 = cv2.rectangle(img, (xb,yb), (x3,y3), (250,0,250),2)
```

```python
ishow(img)
```
