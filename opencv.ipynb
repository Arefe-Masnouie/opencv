```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("lastofus.jpg")
```

```python
img.shape
```

```python
img[0][0]
```

```python
cv2.imshow('ax1', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(img)
```

```python
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img2)
```

```python
img_cr = img2[200:900,500:1100]
```

```python
plt.imshow(img_cr)
```

```python
size = (100,100)
img_resize = cv2.resize(img_cr, size, cv2.INTER_AREA)
```

```python
plt.imshow(img_resize)
```

```python
img_gray = cv2.cvtColor(img_cr, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(img_gray)
```

```python
cv2.imshow('ax1', img_gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
_,img_tr = cv2.threshold(img_cr, 100,255,0)
```

```python
cv2.imshow('ax1', img_tr)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(img_tr)
```

```python
img_bl = cv2.GaussianBlur(img_cr, (51,51), 0)
```

```python
plt.imshow(img_bl)
```

```python
imgc = img2.copy()
cv2.rectangle(imgc, (500,200),(1000,800), (255,0,0), 3)
```

```python
plt.imshow(imgc)
```

```python
cv2.putText(img2,"The Last Of Us", (1250,1000),cv2.FONT_HERSHEY_COMPLEX, 2,(0,255,0),2)
```

```python
plt.imshow(img2)
```
